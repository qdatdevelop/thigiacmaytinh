```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Đọc ảnh và chuyển sang ảnh xám
image = Image.open('C:/Users/tungi/OneDrive/Documents/src_cv/img/01/animal.jpg').convert('L')

# Chuyển ảnh PIL sang mảng NumPy
image_array = np.array(image)

binary_image = 1 * (image_array < 200)

# Thực hiện opening để loại bỏ nhiễu và nối các đối tượng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((9, 5)))

# Đánh nhãn các đối tượng trong ảnh sau khi đã xử lý
labeled_image, num_features = ndimage.label(opened_image)

# Tạo lưới subplot với 2 dòng và 2 cột
fig, axs = plt.subplots(2, 2, figsize=(20, 20))

# Hiển thị ảnh gốc
axs[0, 0].imshow(image_array, cmap='gray')
axs[0, 0].set_title("Ảnh gốc")

# Hiển thị ảnh nhị phân
axs[0, 1].imshow(binary_image, cmap='gray')
axs[0, 1].set_title("Ảnh nhị phân")

# Hiển thị ảnh sau opening
axs[1, 0].imshow(opened_image, cmap='gray')
axs[1, 0].set_title("Ảnh sau opening")

# Hiển thị ảnh đã đánh nhãn
axs[1, 1].imshow(labeled_image, cmap='nipy_spectral')
axs[1, 1].set_title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show()
```
