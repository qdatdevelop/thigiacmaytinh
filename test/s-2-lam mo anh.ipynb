```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Đọc ảnh từ Pillow
image_path = 'C:/Users/tungi/OneDrive/Documents/src_cv/img/01/1.jpg'
image_pil = Image.open(image_path)

# Chuyển ảnh sang ảnh xám
gray_image_pil = image_pil #.convert('L')

# Chuyển ảnh PIL sang mảng NumPy
image = np.array(gray_image_pil)

# Làm mờ ảnh xám với Gaussian
sigma_a = 0.5
blurred_image_a = ndimage.gaussian_filter(image, sigma=sigma_a, mode="constant")

sigma_b = 1
blurred_image_b = ndimage.gaussian_filter(image, sigma=sigma_b, mode="constant")

sigma_c = 2
blurred_image_c = ndimage.gaussian_filter(image, sigma=sigma_c, mode="constant")

# Tạo subplot với lưới 2x2
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Ảnh xám gốc")

plt.subplot(2, 2, 2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title(f"Ảnh xám mờ - Sigma {sigma_a}")

plt.subplot(2, 2, 3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title(f"Ảnh xám mờ - Sigma {sigma_b}")

plt.subplot(2, 2, 4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title(f"Ảnh xám mờ - Sigma {sigma_c}")

plt.tight_layout()
plt.show()
```
