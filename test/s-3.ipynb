```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Đọc ảnh và chuyển sang ảnh xám
image = Image.open('C:/Users/tungi/OneDrive/Documents/src_cv/img/01/face1.jpg').convert('L')

# Chuyển ảnh PIL sang mảng NumPy
image_array = np.array(image)

# Tính gradient theo hướng x (đạo hàm riêng theo x)
gradient_x = ndimage.sobel(image_array, axis=1)

# Thiết lập ngưỡng
threshold = 200  # Điều chỉnh giá trị ngưỡng theo mong muốn

# Tạo ảnh nhị phân từ gradient
binary_image = (np.abs(gradient_x) > threshold).astype(np.uint8) * 255

# Hiển thị ảnh gốc và đường biên
plt.subplot(121)
plt.imshow(image_array, cmap='gray')
plt.title("Ảnh gốc")

plt.subplot(122)
plt.imshow(binary_image, cmap='gray')
plt.title("Đường biên")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Đọc ảnh và chuyển sang ảnh xám
image = Image.open('C:/Users/tungi/OneDrive/Documents/src_cv/img/01/face1.jpg').convert('L')

# Chuyển ảnh PIL sang mảng NumPy
image_array = np.array(image)

# Tính gradient theo hướng x (đạo hàm riêng theo x) bằng mặt nạ Prewitt x
gradient_x = ndimage.convolve(image_array, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng y (đạo hàm riêng theo y) bằng mặt nạ Prewitt y
gradient_y = ndimage.convolve(image_array, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tính biên độ gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(6, 10))

# Hiển thị ảnh gốc, gradient theo hướng x và biên độ gradient
axs[0].imshow(image_array, cmap='gray')
axs[0].set_title("Ảnh gốc")

axs[1].imshow(gradient_x, cmap='gray')
axs[1].set_title("Gradient theo hướng x")

axs[2].imshow(gradient_magnitude, cmap='gray')
axs[2].set_title("Biên độ gradient")

plt.tight_layout()
plt.show()
```
