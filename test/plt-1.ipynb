```python
# Thay đổi màu sắc
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.plot(x, y, color='red')  # Thay đổi màu thành đỏ
plt.show()
```

```python
# Thêm tiêu đề cho biểu đồ
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.plot(x, y, color='red')  # Thay đổi màu thành đỏ
plt.title('Biểu đồ ví dụ với màu đỏ')  # Thêm tiêu đề
plt.show()
```

```python
# Tắt hệ trục tọa độ
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.plot(x, y, color='red')  # Thay đổi màu thành đỏ
plt.title('Biểu đồ ví dụ với màu đỏ')  # Thêm tiêu đề
# Tắt hệ trục tọa độ
plt.axis('off')  # Tắt hệ trục tọa độ
plt.show()
```

```python
# Vẽ trên hình ảnh
# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path)
x = [100, 100, 400, 400]
y = [200, 300, 200, 300]

plt.imshow(im) # hiển thị hình ảnh
plt.plot(x,y,'r*') # đánh dấu các điểm với dấu sao màu đỏ
#plt.plot(x,y,'go-') # đánh dấu các điểm với dấu tròn màu xanh
plt.plot(x,y,'ks:') # đánh dấu các điểm với hình vuông đen
#plt.plot(x, y, color='white') # kết nối các điểm
plt.title('Vẽ trên ảnh') # tiêu đề
plt.show()
```

```python
# Vẽ trên hình ảnh
# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path)
x = [100, 100, 400, 400]
y = [200, 300, 200, 300]

plt.imshow(im) # hiển thị hình ảnh

plt.plot(x, y, color='white') # kết nối các điểm
plt.title('Vẽ trên ảnh') # tiêu đề
plt.show()
```
