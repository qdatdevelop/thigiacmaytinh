```python
# Import thư viện
from PIL import Image
from matplotlib import *
import matplotlib.pyplot as plt
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/tungi/OneDrive/Documents/src_cv/img/01/1.jpg'
# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")
plt.imshow(im)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Chuyển ảnh thành mảng NumPy (đã thực hiện trong mã trước đó)
im = np.array(im)

# Tạo một hình vẽ mới
plt.figure()

# Tạo biểu đồ histogram cho pixel values của ảnh (im)
# im.flatten() làm phẳng mảng pixel thành mảng 1D
# 128 là số lượng bins (ngăn) cho biểu đồ
plt.hist(im.flatten(), bins=128)

# Đặt tiêu đề cho biểu đồ
plt.title("Biểu đồ Histogram của Ảnh")

# Đặt nhãn trục x
plt.xlabel("Giá trị Pixel")

# Đặt nhãn trục y
plt.ylabel("Số lượng")

# Hiển thị biểu đồ histogram
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im, origin="image")
plt.axis('equal')
plt.show()
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/tungi/OneDrive/Documents/src_cv/img/01/face1.jpg'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(im)

# Tạo một hình vẽ mới
plt.figure()

# Đặt màu sắc của hình vẽ thành màu xám
plt.gray()

# Vẽ biểu đồ đường viền (contour) của ảnh với nguồn gốc tại điểm gốc của ảnh
plt.contour(im, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# Hiển thị hình vẽ
plt.show()
```
