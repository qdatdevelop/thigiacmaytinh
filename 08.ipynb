```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Values')  
plt.title('Biểu đồ') 
plt.xlabel('x') 
plt.ylabel('y')  
plt.legend()  
plt.show()  
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  
plt.title('Biểu đồ hộp') 
plt.ylabel('Value')  
plt.show()  
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)  
plt.title('Biểu đồ violin')  
plt.ylabel('Value')  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

cate = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(cate)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar')
plt.show()
```
